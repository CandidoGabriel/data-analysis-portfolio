```python
# Verifica se há valores nulos (buracos no banco)
print(df.isnull().sum())

# Mostra as estatísticas básicas (média, mínimo, máximo, etc.)
print(df.describe())
```

Por essa parte, eu busquei fazer a limpeza de dados, mas o banco de dados já está bem feito, limpo, e pronto para fazer análise. O que é realmente interessante. Em um futuro momento pretendo pegar um banco de dados bruto para fazer a limpeza o mesmo para desenvolver habilidade.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gaming_hours', y='grades', data=df, alpha=0.5, color='purple')

# Adicionando uma linha de tendência (o "caminho" dos dados)
sns.regplot(x='gaming_hours', y='grades', data=df, scatter=False, color='red')

plt.title('Relação entre Horas de Jogo e Notas')
plt.xlabel('Horas de Jogo')
plt.ylabel('Notas')
plt.show()
```

Primeiramente, ver aquilo que é tão discutido que, quanto mais hora de jogo você joga, menor suas notas. O que de certa maneira, é verdade mas, não é o principal vetor que venha a influenciar. o desempenho geral. Apesar de realmente ter uma descida, ele é muito esparso, o que leva a entender que, não é o eixo principal relacionado com o desempenho. Pelo que podes perceber.

```python
plt.figure(figsize=(10, 6))
# Regplot para ver a inclinação da tendência diretamente
sns.regplot(x='gaming_hours', y='study_hours', data=df, color='green', scatter_kws={'alpha':0.3})

plt.title('Correlação: Horas de Jogo vs Horas de Estudo')
plt.xlabel('Horas de Jogo')
plt.ylabel('Horas de Estudo')
plt.show()
```

Aqui, houve uma busca pra ver se as horas de jogo e a horas toda impactaria uma na outra, mas, aparentemente não. O gráfico está bem distribuído, o que ocasiona uma variedade muito grande pra dizer que é algo que esteja impactando diretamente tanto em um quanto em outro.

```python
plt.figure(figsize=(8, 6))
# Boxplot para comparar categorias (Low, Medium, High)
sns.boxplot(x='stress_level', y='gaming_hours', data=df, order=['Low', 'Medium', 'High'], palette='viridis')

plt.title('Nível de Estresse vs Horas de Jogo')
plt.xlabel('Nível de Estresse')
plt.ylabel('Horas de Jogo')
plt.show()
```

Aqui está um gráfico que, busquei ver a relação de estresse e horas de jogo. E pude perceber principalmente que acima de seis horas o nível de estresse é baixo, o que acasionalmente, pra quem está passando sete a oito horas, é um tempo consideravelmente alto. Então, é uma sessão dedicada pra aquilo. E, sem pressão nenhuma, mas aqueles que estão entre duas a seis horas, que têm alto nível de estresse, dá pra se perceber talvez, que sejam sessões relativamente longas e que possam ter sido interrompidas por não ter tanto tempo livre, assim como aqueles que estão dedicando até sete horas, oito horas. E tem aqueles que estão no meio termo, de nível de estresse, entre duas a quatro horas, que ocasionalmente aparenta ser algo mais equilibrado, quanto a outras atividades. Apesar da que está mostrando apenas o nível de estresse de jogos, mas buscando fazer uma relação com outros parâmetros que tem no banco de dados. Quem ainda pretendo ver.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='gaming_hours', y='sleep_hours', data=df, color='orange', scatter_kws={'alpha':0.3})
plt.title('Impacto do Jogo no Tempo de Sono')
plt.xlabel('Horas de Jogo')
plt.ylabel('Horas de Sono')
plt.show()
```

O que foi buscado ver se os usuários de jogos impacta na horas do sono, e se isso daí acaba influenciando também em perda de desempenho. O que está muito disperso, e potencialmente não tem muita ligação como algo que impacta de maneira negativa no desempenho.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gaming_hours', y='device_usage', data=df, color='blue', alpha=0.5)
plt.title('Horas de Jogo vs Uso de Dispositivos (Device Usage)')
plt.xlabel('Horas de Jogo')
plt.ylabel('Uso de Dispositivos')
plt.show()
```

Que é interessante, como tem uma variação de, pelo menos quatro horas de uso de tela. E como isso daí também está relacionado ao uso de dispositivos. Obviamente, se vai jogar vai jogar num dispositivo, e desses dispositivos no caso pode ser celular, pode ser computador, e talvez esteja sendo redundante, talvez, mas, ainda assim, tem uma boa ligação do que está sendo feito, principalmente com relação a um certo equilíbrio. que acontece entre os dois.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='device_usage', y='study_hours', data=df, color='orange', scatter_kws={'alpha':0.3})
plt.title('Impacto do Uso de Dispositivos nas Horas de Estudo')
plt.xlabel('Uso de Dispositivos (Horas)')
plt.ylabel('Horas de Estudo')
plt.show()
```

Isso. Hoje eu não busco é de ver se o impacto de hoje dispositivo está relacionado com, o impacto de horas. E, de certa maneira, não muito. Apesar que, dá pra ver que tem uma tendência clara aos dados positivo por de seis a dez horas. Este é relativamente interessante, mesmo pra pessoas que estudam por bastante horas. E, dá pra fazer uma ligação bem interessante. Tanto os dispositivos, horas de estudo, horas de jogo, não impactando na horas de sono, e eu corro regionalmente, o nível de estresse. Talvez avendo um certa necessidade de prioridade e distribuição de tempo.

```python
plt.figure(figsize=(8, 6))
# Boxplot para comparar a distribuição de horas de estudo por nível de estresse
sns.boxplot(x='stress_level', y='study_hours', data=df, order=['Low', 'Medium', 'High'], palette='coolwarm')
plt.title('Horas de Estudo por Nível de Estresse')
plt.xlabel('Nível de Estresse')
plt.ylabel('Horas de Estudo')
plt.show()
```

e é algo interessante, que aqueles que estudam de oito a dez horas têm o estresse maior, e esse tempo de estudo dedicado de oito a dez horas impacta sim, o que eles vão dedicar pra jogar, o que eles vão dedicar pra ver tela, e até mesmo acaba combinando os dois, o que é realmente interessante, pois você acaba vendo pessoas que têm o estresse maior, e como elas vão fazer isso, esse grupo em específico pretendo analisálo de maneira mais profunda, a fim de ver algumas coisas que possam comprovar o que eu acabei vendo e uma análise anterior.

```python
# Filtrando alunos com estresse alto
high_stress = df[df['stress_level'] == 'High']

# Média de notas desse grupo vs média geral
print(f"Média de notas (Geral): {df['grades'].mean():.2f}")
print(f"Média de notas (Estresse Alto): {high_stress['grades'].mean():.2f}")
```

Primeiramente fui ver qual é a média desse grupo, se eles têm realmente a maior nota ou se eles, Estão com notas baixas e tentando se recuperar. mas pela média alta. e bem distante do manejo geral, dá pra ver que eles realmente estão se dedicando a ter o maior desempenho, e pagando o preço tendo maior estresse.

```python
# Filtra estudantes com Estresse Alto E notas acima da média geral
high_performers = df[(df['stress_level'] == 'High') & (df['grades'] > 66.18)]

# Calcula a média de horas de jogo desse grupo específico
media_jogo_high_perf = high_performers['gaming_hours'].mean()

print(f"Número de alunos neste perfil: {len(high_performers)}")
print(f"Média de horas de jogo deste grupo: {media_jogo_high_perf:.2f} horas")

# Visualizando a distribuição de horas de jogo desse grupo
plt.figure(figsize=(8, 5))
sns.histplot(high_performers['gaming_hours'], kde=True, color='red')
plt.title('Distribuição de Horas de Jogo: Perfil "Alto Estresse / Alto Desempenho"')
plt.xlabel('Horas de Jogo')
plt.show()
```

Dá pra ver que essas pessoas que têm um alto desempenho em alto nível de estresse, apresentam uma distribuição relativamente interessante na horas de jogo, desde zero até às seis horas. E essa distribuição equilibrada e uniforme, dá pra ver que, horas de jogo não é o que impacta necessariamente. no no desempenho. Talvez seja algo relacionado a buscar, equilibrar, tempo que está estudando, tempo que está. uma atividade recreativa.

```python
# Reutilizando o grupo high_performers
# Calcular as médias de tempo de uso para este grupo
resumo_tempo = high_performers[['study_hours', 'gaming_hours', 'device_usage']].mean()

print("Distribuição média do tempo do grupo de Alta Performance:")
print(resumo_tempo)

# Visualizando essa distribuição (Gráfico de Barras)
resumo_tempo.plot(kind='bar', color=['green', 'purple', 'blue'], figsize=(8, 5))
plt.title('Distribuição Média de Tempo (Grupo Alta Performance)')
plt.ylabel('Horas Médias')
plt.xticks(rotation=45)
plt.show()
```

Aparentemente dá pra ver que uso dos dispositivos, é onde acontece os jogos e parte dos estudos. E é interessante pensar nisso que ele não engloba os dois no geral, mas uma parte de um, grande parte de um, nele todo, e uma parte do outro. Então dá pra ver que parte dos estudos ele acontece fora dos de telas, aonde dá pra se considerar que se reforça o alto desempenho.
