```python
import pandas as pd
import numpy as np

# Carregar o dataset
df = pd.read_csv('SteamGames.csv')

# Visualizar as primeiras linhas e informações das colunas
print(df.info())
df.head(6)
```

Banco de dados acessado! agora o objetivo é limpar ele para a analise, a meta é ter uma analise exploratoria para ver o estado do mercado entendendo o que esta em alta apartir de uma imagem de alta resolução baseado nos dados disponiveis e filtrados com apenas conteudo relevante para esse fim.

```python
# Remove o símbolo de cifrão, remove espaços e converte para número decimal (float)
df['price_clean'] = df['price'].astype(str).str.replace('$', '', regex=False)
df['price_clean'] = pd.to_numeric(df['price_clean'], errors='coerce')

# Preencher jogos gratuitos ou erros com 0
df['price_clean'] = df['price_clean'].fillna(0)
```

```python
# Converter para o formato de data do Pandas e extrair o Ano
df['ReleaseYear'] = pd.to_datetime(df['ReleaseDate'], errors='coerce').dt.year

# Filtrar o DataFrame para remover anos muito antigos ou nulos, focando no que importa
df = df[df['ReleaseYear'] >= 2015]
```

```python
# Garantir que as colunas de review sejam numéricas
df['PositiveReview'] = pd.to_numeric(df['PositiveReview'], errors='coerce').fillna(0)
df['NegativeReview'] = pd.to_numeric(df['NegativeReview'], errors='coerce').fillna(0)

# Criar coluna de Total de Avaliações (Volumetria de jogadores)
df['TotalReviews'] = df['PositiveReview'] + df['NegativeReview']

# Criar a Taxa de Aprovação (Apenas para jogos que possuem avaliações)
df['ApprovalRate'] = np.where(df['TotalReviews'] > 0, df['PositiveReview'] / df['TotalReviews'], 0)
```

```python
colunas_para_manter = [
    'Appid', 'Name', 'Type', 'ReleaseYear', 'Genres', 
    'Developers', 'Publishers', 'price_clean', 'Tags', 
    'ReviewScore', 'PositiveReview', 'NegativeReview', 'TotalReviews', 'ApprovalRate'
]

df_limpo = df[colunas_para_manter].copy()
```

Filtrar as colunas que serão usando discartando aquelas que não são relevantes para o objetivo da analise.

```python
# Salvar o novo arquivo limpo
df_limpo.to_csv('SteamGames_Limpo.csv', index=False)
print("Banco de dados limpo com sucesso! Pronto para análise.")
```
